In this notebook we are going to make the models vote for the correct target! To do so we are going to train a logistic regression on traindata using the predictions of the different models as independent variables and the column "target" as dependent variables.

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
traindata = pd.read_csv("VOTE_train.csv") 
submitdata = pd.read_csv("VOTE_submit.csv")
```

```python
traindata['predicted_target']=str('0')
traindata
```

```python
submitdata
```

```python
# define X_train and y_train
X_train = traindata[['KNN_prediction','NB_prediction', 'SVC_prediction', 'RF_prediction', 'LR_prediction']]
y_train = traindata['target']


X_submit=submitdata[['KNN_prediction','NB_prediction', 'SVC_prediction', 'RF_prediction', 'LR_prediction']]
```

```python
# train logistic regression on the models' predictions
lr = LogisticRegression()
lr.fit(X_train, y_train)
traindata[['predicted_target']]=lr.predict(X_train)

# now let's vote!
submitdata[['target']]=lr.predict(X_submit)
```

```python
# display misclassified tweets

pd.set_option('display.max_colwidth', -1)

incorrectly_classified = traindata[traindata['predicted_target']!=traindata['target']]
incorrectly_classified
```
