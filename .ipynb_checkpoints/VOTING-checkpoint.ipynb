```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
traindata = pd.read_csv("VOTE_train.csv") 
submitdata = pd.read_csv("VOTE_submit.csv")
```

```python
traindata['predicted_target']=str('0')
traindata
```

```python
submitdata
```

```python
X_train = traindata[['KNN_prediction', 'NB_prediction', 'SVM_prediction', 'RF_prediction', 'LR_prediction']]
y_train = traindata['target']

X_submit=submitdata[['KNN_prediction', 'NB_prediction', 'SVM_prediction', 'RF_prediction', 'LR_prediction']]
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
submitdata[['target']]=lr.predict(X_submit)
```

```python
submitdata.head()
```

```python
to_submit=submitdata[['id', 'target']]
to_submit
```

```python
# Save to csv
to_submit.to_csv("MEGA_SUBMIT.csv", index=False, encoding = 'utf-8')
```
